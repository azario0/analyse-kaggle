# Analyse des données et visualisation de 'data.csv'

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Chargement des données

```python
df = pd.read_csv('data.csv')
```

### Inspection des données

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

## Statistiques descriptives et visualisation

### Distribution des prix

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribution du prix')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['offer_price'], bins=30, kde=True)
plt.title('Distribution du prix promotionnel')
plt.show()
```

### Relation entre le prix et le prix promotionnel

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='price', y='offer_price', data=df)
plt.title('Prix vs Prix promotionnel')
plt.show()
```

## Analyse des variables catégorielles

### Analyse des marques

```python
print(df['brand'].nunique())
```

```python
brand_counts = df['brand'].value_counts()
print(brand_counts.head(10))
```

```python
plt.figure(figsize=(12,6))
brand_counts.head(10).plot(kind='bar')
plt.title('Top 10 des marques par nombre')
plt.xlabel('Marque')
plt.ylabel('Nombre')
plt.show()
```

### Analyse des couleurs

```python
color_counts = df['color'].value_counts()
print(color_counts.head(10))
```

```python
plt.figure(figsize=(10,6))
color_counts.head(10).plot(kind='bar')
plt.title('Top 10 des couleurs par nombre')
plt.xlabel('Couleur')
plt.ylabel('Nombre')
plt.show()
```

### Prix moyen par marque

```python
top_brands = brand_counts.head(10).index
df_top_brands = df[df['brand'].isin(top_brands)]
plt.figure(figsize=(12,6))
df_top_brands.groupby('brand')['price'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Prix moyen des top 10 marques')
plt.xlabel('Marque')
plt.ylabel('Prix moyen')
plt.show()
```

## Analyse des remises

```python
df['discount_amount'] = df['price'] - df['offer_price']
df['discount_percentage'] = (df['discount_amount'] / df['price']) * 100
```

### Distribution du pourcentage de remise

```python
plt.figure(figsize=(10,6))
sns.histplot(df['discount_percentage'], bins=30, kde=True)
plt.title('Distribution du pourcentage de remise')
plt.show()
```

## Analyse de corrélation

```python
df["size"] = pd.to_numeric(df["size"], errors="coerce")
df["price"] = pd.to_numeric(df["price"], errors="coerce")
df["offer_price"] = pd.to_numeric(df["offer_price"], errors="coerce")

cleaned_df = df.dropna()

numerical_df = cleaned_df[["size", "price", "offer_price"]]

correlation_matrix = numerical_df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de corrélation")
plt.show()
```
